```python
prodCats = ('beverages', 'phoneAcces','toiletries','pastry', 'cosmetics') #product Categories

prodDict = {'beverages':['Chocolate', 'Drinks', 'Coffee', 'Tea', 'Soy drinks', 'Pop', 'Soda'], 
            'phoneAcces':['Cover case', 'Earpieces', 'Safe Quards'],
            'toiletries':['Toilet paper', 'Bathing soap', 'Scrubs', 'Body crème', 'Shampoo'],
            'pastry':['Pizza', 'Burgers', 'Donuts', 'Muffins', 'Cheesecakes'], 
            'cosmetics':['Perfumes', 'Vanishes', 'Nail polish', 'Deodorants', 'Face scrubs']
           } #product items per categories

bevInventDict = {'Chocolate':100, 'Drinks':4, 'Coffee':150, 'Tea':125, 'Soy drinks':164, 'Pop':80, 'Soda':100}
pAccInventDict = {'Cover case': 180, 'Earpieces': 200, 'Screen quards' : 250}
toiInventDict = {'Toilet paper': 250, 'Bathing soap': 230, 'Scrubs':200, 'Body crème':300, 'Shampoo':150}
pastInventDict = {'Pizza': 20, 'Burgers' : 25, 'Donuts' : 30, 'Muffins': 15, 'Cheesecakes': 2}
cosInventDict = {'Perfumes': 105, 'Vanishes': 89, 'Nail polish': 90, 'Deodorants': 100, 'Face scrubs': 70}
inventoryList = ['bevInventDict', 'pAccInventDict','toiInventDict','pastInventDict','cosInventDict'] #product items inventory
```

```python
def checkinventory(prod):
    
    for k, v in prodDict.items(): #checking item inventory and alerting the owner if items needs to be restock
        
        category = k
        prodList = v
        prodInvent = ''
        
        if prod in prodList:
            if category == 'beverages':
                prodInvent = bevInventDict[prod]
                if prodInvent < 5:
                    print(f'There is need to restock {prod} only {prodInvent} remaining')
                else:
                    return prodInvent
                return prodInvent
        
            elif category == 'phoneAcces':
                prodInvent = pAccInventDict[prod]
                if prodInvent < 5:
                    print(f'There is need to restock {prod} only {prodInvent} remaining')
                else:
                    return prodInvent
                return prodInvent
                
            elif category == 'toiletries':
                prodInvent = toiInventDict[prod]
                if prodInvent < 5:
                    print(f'There is need to restock {prod} only {prodInvent} remaining')
                else:
                    return prodInvent
                return prodInvent
            
            elif category == 'pastry':
                prodInvent = pastInventDict[prod]
                if prodInvent < 5:
                    print(f'There is need to restock {prod} only {prodInvent} remaining')
                else:
                    return prodInvent
                return prodInvent
        
            elif category == 'cosmetics':
                prodInvent = cosInventDict[prod]
                if prodInvent < 5:
                    print(f'There is need to restock {prod} only {prodInvent} remaining')
                else:    
                    return prodInvent
                return prodInvent
               
    else:
        return  f'{prod}  not found in inventory'
    
prodCheck = input('Enter product: ')
prodCount = checkinventory(prodCheck)
print(prodCount)
```

```python
# Code to update inventory


# Flow:

1. check if product is in any dictionary
2. if found, reference the coresponding invntory dictionary and increment quantity by the new amout
3. if not found, prompt user to chose a category to add the product
    update the corresponding inventory dictionay with product and inventory
```

```python
def inventoryUpdater(prod, qty):    #checking item inventory for restocking and allowing the owner to restock
    for k, v in prodDict.items():
        
        category = k
        prodList = v
        
        if prod in prodList:
            if category == 'beverages':
                oldQty = bevInventDict[prod]
                bevInventDict[prod] += qty 
                print('update successful')
                return [oldQty, bevInventDict[prod]]
            
            elif category == 'phoneAcces':
                oldQty = pAccInventDict[prod]
                pAccInventDict[prod] += qty 
                print('update successful')
                return [oldQty,pAccInventDict[prod]]

            elif category == 'toiletries':
                oldQty = toiInventDict[prod]
                toiInventDict[prod] += qty 
                print('update successful')
                return [oldQty, toiInventDict[prod]]

            elif category == 'pastry':
                oldQty = pastInventDict[prod]                       
                pastInventDict[prod] += qty 
                print('update successful')
                return [oldQty, pastInventDict[prod]]

                
            elif category == 'cosmetics':
                oldQty = cosInventDict[prod]
                cosInventDict[prod] += qty 
                print('update successful')
                return [oldQty, cosInventDict[prod]]

    else:
        return  [f'{prod}  not found in inventory, do you want to add as new product?', '...']
    
    
newInvent = input('specify item :' )
invetQty = int(input('specify qty :' ))

oldQty, newQty= inventoryUpdater(newInvent, invetQty)

if newQty != '...':
    print(f'{newInvent} inventory updated from {oldQty} to  {newQty}')
else:
    print('not found in inventory, do you want to add as new product?')
```
